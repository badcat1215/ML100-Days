# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
#作業:寫出並印出 ReLU & dReLU 一階導數
```

```python
#True = 1 , False = 0
def ReLU(x):
    return abs(x) * (x > 0)

def dReLU(x):
    return (1 * (x > 0))

# generate an array from start and stop value
# with requested number of elements
x = np.linspace(-10,10,100)

# prepare the plot, associate the color red or blue and the label
plt.plot(x, ReLU(x), 'b')
plt.plot(x, dReLU(x), 'r')

# draw the grid line in background.
plt.grid()

# title
plt.title('ReLU Function')

# write the ReLU formula
plt.text(-8, 5, r'$ReLU(x)=max(0,x)$',fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

# create the graph
plt.show()
```
